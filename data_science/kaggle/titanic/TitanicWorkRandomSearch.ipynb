```python
max_titanic_features = 116

# try one-hot encoding on categorical variables - DONE
# missing values NaN boolean column - DONE
# try XGBoost - LATER

# build classifiers
clf = RandomForestClassifier(n_estimators=2000)  # change max_depth; try defaults w/2000 estimators
clf_params = {"max_depth": [3, 7, 13, 19, 31, 47, 71, 97], "max_features": sp_randint(1, max_titanic_features),
              "min_samples_split": sp_randint(2, max_titanic_features), "bootstrap": [True, False],
              "criterion": ["gini", "entropy"]}

gbc = GradientBoostingClassifier(n_estimators=1000)
gbc_params = {"max_depth": [3, 4, 5, 6, 7], "max_features": sp_randint(1, max_titanic_features),
              "min_samples_split": sp_randint(2, 10), # need at least this to split
              "criterion": ['friedman_mse', 'mse', 'mae'], "learning_rate" : [0.01, 0.03, 0.1, 0.3, 1.0],
              "loss" : ['deviance', 'exponential']}

knc = KNeighborsClassifier()
knc_params = {"n_neighbors": [3, 5, 7, 9],
              "weights": ['uniform', 'distance'],
              "algorithm": ['auto', 'ball_tree', 'kd_tree', 'brute'],
              "metric": ['euclidean', 'manhattan', 'chebyshev', 'minkowski']}

model = clf
param_dist = clf_params

# Utility function to report best scores
def report(results, n_top=3):
    for i in range(1, n_top + 1):
        candidates = np.flatnonzero(results['rank_test_score'] == i)
        for candidate in candidates:
            print("Model with rank: {0}".format(i))
            print("Mean validation score: {0:.3f} (std: {1:.3f})".format(
                  results['mean_test_score'][candidate],
                  results['std_test_score'][candidate]))
            print("Parameters: {0}".format(results['params'][candidate]))
            print("")


# specify parameters and distributions to sample from

# run randomized search
n_iter_search = 1000  # increased from 100
random_search = RandomizedSearchCV(model, param_distributions=param_dist, n_iter=n_iter_search, cv=5)

start = time()
random_search.fit(X_train, Y_train)
print("RandomizedSearchCV took %.2f seconds for %d candidates"
      " parameter settings." % ((time() - start), n_iter_search))
report(random_search.cv_results_)
```
